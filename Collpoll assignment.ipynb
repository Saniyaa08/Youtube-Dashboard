```python
import pandas as pd

# Load the provided dataset
data = pd.read_csv(r"C:\Users\DELL\Downloads\Global YouTube Statistics.csv")
data
```

```python
# Check the data types of 'subscribers' and 'video views' columns
print(data['subscribers'].dtype)
print(data['video views'].dtype)
```

```python
# Check for missing values in each column
missing_values = data.isna().sum()

# Display the columns with missing values and their counts
columns_with_missing_values = missing_values[missing_values > 0]
print(columns_with_missing_values)
```

```python
# List of columns with missing values
columns_with_missing_values = data.columns[data.isnull().any()].tolist()
```

```python
# Handling missing values for numeric columns (imputation with mean)
numeric_columns = ['video_views_rank', 'country_rank', 'channel_type_rank', 'video_views_for_the_last_30_days',
                   'subscribers_for_last_30_days', 'created_year', 'created_month', 'created_date',
                   'Gross tertiary education enrollment (%)', 'Population', 'Unemployment rate',
                   'Urban_population', 'Latitude', 'Longitude']
```

```python
# Convert columns to numeric and then fill missing values with the mean
for column in numeric_columns:
    data[column] = pd.to_numeric(data[column], errors='coerce')  # Convert to numeric, coerce errors to NaN
    data[column].fillna(data[column].mean(), inplace=True)
```

```python
# Handling missing values for categorical columns (fill with a placeholder category)
categorical_columns = ['category', 'Country', 'Abbreviation', 'channel_type']

for column in categorical_columns:
    data[column] = data[column].fillna('Unknown')
```

```python
# Display the columns with missing values after handling
missing_values = data.isna().sum()
columns_with_missing_values = missing_values[missing_values > 0]
print(columns_with_missing_values)
```

```python
# Check for missing values in the 'created_month' column
missing_created_month = data['created_month'].isna().sum()
```

```python
# If there are missing values, you can fill them with a default value (e.g., 0 for unknown month)
if missing_created_month > 0:
    data['created_month'].fillna(0, inplace=True)
```

```python
# Verify that the missing values in 'created_month' have been filled
missing_values = data.isna().sum()
columns_with_missing_values = missing_values[missing_values > 0]
print(columns_with_missing_values)
```

```python
# Define the layout of the dashboard
app.layout = html.Div([
    html.H1("YouTube Channel Rankings Dashboard", style={'textAlign': 'center'}),
    
    # Dropdown for selecting metrics
    dcc.Dropdown(
        id='metrics-dropdown',
        options=[
            {'label': 'Subscribers', 'value': 'subscribers'},
            {'label': 'Video Views', 'value': 'video views'},
            # Add more metrics as needed
        ],
        value='subscribers',
        style={'width': '50%'}
    ),
    
    # Search bar
    dcc.Input(
        id='search-bar',
        type='text',
        placeholder='Search for a Youtuber...',
        style={'width': '50%'}
    ),
    
    # Bar chart for rankings
    dcc.Graph(id='rankings-chart'),

    # Additional visualizations and components can be added here
])

# Callback to update the rankings chart based on user inputs
@app.callback(
    Output('rankings-chart', 'figure'),
    Input('metrics-dropdown', 'value'),
    Input('search-bar', 'value')
)
def update_rankings_chart(selected_metric, search_query):
    # Your code to update the chart based on user inputs goes here
    # You can filter and sort the data based on the selected_metric and search_query
    
    # Example: Create a bar chart of rankings
    if selected_metric == 'subscribers':
        sorted_data = data.sort_values(by='subscribers', ascending=False)
    else:
        sorted_data = data.sort_values(by='video views', ascending=False)
    
    if search_query:
        filtered_data = sorted_data[sorted_data['Youtuber'].str.contains(search_query, case=False)]
    else:
        filtered_data = sorted_data
    
    fig = px.bar(filtered_data, x='Youtuber', y=selected_metric, title=f'Channel Rankings by {selected_metric}')
    
    return fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
